```python

import matplotlib.pyplot as plt

brands = ['Lamborghini', 'Bentley', 'Porsche', 'Jaguar', 'Land', 
          'Mini', 'Audi', 'Mercedes-Benz', 'BMW', 'Isuzu']
avg_prices = [4.8, 4.1, 3.7, 3.5, 3.3, 3.2, 3.1, 3.0, 3.0, 2.9] 
plt.bar(brands, avg_prices, color='darkblue', edgecolor='yellow')
plt.title("Brands Price")
plt.xlabel("Brand")
plt.ylabel("Average Price (Cr)")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV directly
df = pd.read_csv("used_cars_data.csv")

# Simple Price histogram
plt.hist(df["Price"].dropna(), bins=20, color="skyblue", edgecolor="Yellow")

plt.xlabel("Price (Cr)")
plt.ylabel("Number of Cars")
plt.title("Histogram")
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("used_cars_data.csv")
df_clean = df[["Year", "Price"]].dropna()
plt.figure(figsize=(8, 6))
plt.scatter(df_clean["Year"], df_clean["Price"], color="red", edgecolor="Yellow", alpha=0.7)

plt.xlabel("Year")
plt.ylabel("Price (Cr)")
plt.title("Scatter Plot: Price VS Year")
plt.grid(True)
plt.show()
```
